```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
df.drop(columns=['User ID'],inplace=True)
```

```python
df
```

```python
df['Gender'] = df['Gender'].replace({'Male':1,'Female':0})
```

```python
df
```

```python
x = df.drop(columns=['Purchased'])
y = df['Purchased']
```

```python
x.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x_normalize = StandardScaler()
x_normalize = x_normalize.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_normalize, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train, y_train)
```

```python
y_pred = log_model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f'Accurect = {accuracy*100}%')
```

```python
import joblib
filename = 'logistic_regression_model.sav'
joblib.dump(log_model,filename)
```

```python
load_model = joblib.load('/content/logistic_regression_model.sav')
```

```python
y_pred_load = load_model.predict(x_test)
```

```python
acc_loaded = accuracy_score(y_test, y_pred_load)
print(f'Accurect = {acc_loaded*100}%')
```

```python
print(f"Accuracy of loaded model = {acc_loaded*100}%")
print("Predictions of first 10 samples : ")
print(load_model.predict(x_test[:10]))
```

```python
def display_gender_predictions(predictions):
    purchased_map = {0:'Yes', 1:'No'}
    print("gender prediction(0:Yes),(1:'No')")
    for prediction in predictions:
        print(f"Prediction: {purchased_map[prediction]}")
predictions_for_display = load_model.predict(x_test[:10])
display_gender_predictions(predictions_for_display)
```
