```python
def show_classes_samples(dataset,num_samples=5):
    class_examples = {class_name: [] for class_name in dataset.classes}
    loader = DataLoader(dataset, batch_size=32,shuffle=True)
    for images, labels in loader:
        for img, label in zip(images,labels):
            class_name = dataset.classes[label]
            if len(class_examples[class_name]) < num_samples:
                class_examples[class_name].append(img)

        if all(len(v) >= num_samples  for v in class_examples.values()):
            break
    n_classes = len(dataset.classes)
    fig, axes = plt.subplots(
        nrows=n_classes,
        ncols=num_samples,
        figsize=(15, 2.5 * n_classes),
        squeeze=False,
  )

    for row_idx, class_name in enumerate(dataset.classes):
        for col_idx in range(num_samples):
            ax = axes[row_idx, col_idx]
            img = class_examples[class_name][col_idx].numpy().squeeze()
            img = img * 0.5 + 0.5
            ax.imshow(img, cmap="gray")
            ax.set_title(class_name)
            ax.axis("off")
            if  col_idx ==0:
                 ax.set_title(f"{class_name}({len(dataset.targets)} samples)", loc = 'left')
    plt.tight_layout()
    plt.show()
```

```python
def train_model(model, train_loader,val_loader,criterion,optimizers,epochs=5):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model.to(device)
    best_model_wts = model.state_dict()
    best_acc = 0.0

    for epoch in range(epochs):
       model.train()
       running_loss = 0.0
       all_preds, all_labels = [], []

       loop = tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs}')
       for inputs, labels in loop:
           inputs, labels = inputs.to(device), labels.to(device)
           optimizers.zero_grad()
           outputs = model(inputs)
           # Convert labels to one-hot encoding
           labels_one_hot = torch.nn.functional.one_hot(labels, num_classes=outputs.shape[1]).float()
           loss = criterion(outputs, labels_one_hot)
           loss.backward()
           optimizers.step()

           preds=torch.argmax(outputs, dim=1)
           all_preds.extend(preds.cpu().numpy())
           all_labels.extend(labels.cpu().numpy())
           running_loss += loss.item()
           loop.set_postfix(loss=loss.item())
       train_acc = np.mean(np.array(all_preds) == np.array(all_labels))
       print(f"train acuuracy: {train_acc:.4f}")
       print(classification_report(all_labels, all_preds, target_names=dataset.classes))

       model.eval()
       val_preds, val_labess = [], []
       with torch.no_grad():
           for inputs, labels in val_loader:
               inputs, labels = inputs.to(device), labels.to(device)
               outputs = model(inputs)
               preds = torch.argmax(outputs, dim=1)
               val_preds.extend(preds.cpu().numpy())
               val_labess.extend(labels.cpu().numpy())

       val_acc = np.mean(np.array(val_preds) == np.array(val_labess))
       print(f"val acuuracy: {val_acc:.4f}")
       print(classification_report(val_labess, val_preds, target_names=dataset.classes))
       if val_acc > best_acc:
           best_acc = val_acc
           best_model_wts = model.state_dict()
    model.load_state_dict(best_model_wts)
    return model
```

```python
train_size = int(0.8*len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size,val_size])
train_loader = DataLoader(train_dataset,batch_size=32,shuffle=True)
val_loader = DataLoader(val_dataset,batch_size=32,shuffle=False)
```

```python
model = train_model(model,train_loader,val_loader,criterion,optimizers,epochs = 10)
```

```python
from torchvision.models import efficientnet_b0
import torch
import torch.nn as nn
from torchvision import transforms

def load_model_for_prediction(model_path,num_classes=2):
    loaded_model= efficientnet_b0(pretrained=True)

    # Modify the first convolutional layer to accept 1 channel input
    loaded_model.features[0][0] = nn.Conv2d(
      in_channels=1,
      out_channels = loaded_model.features[0][0].out_channels,
      kernel_size = loaded_model.features[0][0].kernel_size,
      stride = loaded_model.features[0][0].stride,
      padding = loaded_model.features[0][0].padding,
    bias = False
)
    in_features = loaded_model.classifier[1].in_features
    loaded_model.classifier[1] = nn.Linear(in_features, num_classes)
    loaded_model.load_state_dict(torch.load(model_path))
    loaded_model.eval()
    return loaded_model

loaded_model = load_model_for_prediction('bone_fracture_model.pth')

predict_transforms= transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.Resize((256,256)),
    transforms.ToTensor(),
    transforms.Normalize(mean = [0.5], std=[0.5])
])
```

```python
from PIL import Image

def predict_image_fracture(image_path, model,dataset_classes):
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    model.to(device)
    image = Image.open("/content/bone-fracture-dataset/Bone fracture dataset/Bone fracture dataset/Dataset/fracture/1.png")
    image = predict_transforms(image).unsqueeze(0).to(device)

    with torch.no_grad():
        outputs = model(image)
        probabilities = torch.softmax(outputs, dim=1)
        predicted_class_index = torch.argmax(probabilities).item()
        predicted_class_name = dataset_classes[predicted_class_index]
        confidence = probabilities[0, predicted_class_index].item()
        return predicted_class_name, confidence
```
